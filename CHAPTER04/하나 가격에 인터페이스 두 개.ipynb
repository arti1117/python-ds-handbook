```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('classic')
```

```python
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
plt.gcf()
```

```python
plt.gca()
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
